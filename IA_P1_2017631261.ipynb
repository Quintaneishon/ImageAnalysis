**Calculating the histograms for the three channels and displays them**

<h3>Image Histogram</h3>
<p>An image histogram is a type of histogram that acts as a graphical representation of the tonal distribution in a digital image. It plots the number of pixels for each tonal value. By looking at the histogram for a specific image a viewer will be able to judge the entire tonal distribution at a glance.</p><br>


<h3>Python Algorithm</h3>

1.   Create a array with 256 zeros.


```
hist_red=np.zeros(256)
```


2.   The tonal histogram depends of the last number in the triple array 0 for Red, 1 for Green, 2 for Blue. So this value never change. 
3.   The value of the pixel is the position of the zeros array, so this value increment on 1 when it repeats.

```
for x in range(height):
  for y in range(width):
    hist_red[pic_arr[x,y,0]]=hist_red[pic_arr[x,y,0]]+1
```

````python
#Get the height and the width
height=pic_arr.shape[0]
width=pic_arr.shape[1]

if channels == 3:
  #RED CHANNEL
  hist_red=np.zeros(256)
  for x in range(height):
    for y in range(width):
      hist_red[pic_arr[x,y,0]]=hist_red[pic_arr[x,y,0]]+1

  #GREEN CHANNEL
  hist_green=np.zeros(256)
  for x in range(height):
    for y in range(width):
      hist_green[pic_arr[x,y,1]]=hist_green[pic_arr[x,y,1]]+1



  #BLUE CHANNEL
  hist_blue=np.zeros(256)
  for x in range(height):
    for y in range(width):
      hist_blue[pic_arr[x,y,2]]=hist_blue[pic_arr[x,y,2]]+1



  fig, axs = plt.subplots(1, 3, figsize=(20,5), sharey=True)
  axs[0].plot(hist_red,color = "r")
  axs[0].set_xlabel("red")
  axs[1].plot(hist_green,color = "g")
  axs[1].set_xlabel("green")
  axs[2].plot(hist_blue,color = "b")
  axs[2].set_xlabel("blue")
  fig.suptitle('Histogram channels')

elif channels == 1:
  hist=np.zeros(256)
  for x in range(height):
    for y in range(width):
      hist[pic_arr[x,y]]=hist[pic_arr[x,y]]+1
  plt.plot(hist,color = "k")
  plt.show()
````

**An alternative way to calculate and display the histograms in the same plot**

Of course, if you are working with a grayscale image you'll get a single blue plot

````python
height=pic_arr.shape[0]
width=pic_arr.shape[1]

if channels == 3:
  histogram=np.zeros((3,256))  
  for i in range(3):
    for x in range(height):
      for y in range(width):
        histogram[i,pic_arr[x,y,i]]=histogram[i,pic_arr[x,y,i]]+1

  plt.plot(histogram[0],color = "r")
  plt.plot(histogram[1],color = "g")
  plt.plot(histogram[2],color = "b")
  plt.show()

elif channels == 1:
  hist=np.zeros(256)
  for x in range(height):
    for y in range(width):
      hist[pic_arr[x,y]]=hist[pic_arr[x,y]]+1
  plt.plot(hist,color = "k")
  plt.show()
````

**A first aproximation for thresholding**

<h3>Thresholding</h3>
The simplest thresholding methods replace each pixel in an image with a black pixel if the image intensity is less than some fixed constant T, or a white pixel if the image intensity is greater than that constant

````python
T=120 #The Threshold

#color
if channels == 3:
  pic_arr2=pic_arr.copy()
  for channel in range(0,3):   #FOR EVERY CHANNEL (R,G,B)
      for i in range(0,height):   #FOR EVERY ROW   
          for j in range(0,width):   #FOR EVERY COLUMN
              if pic_arr2[i,j,channel]<T: #IF THE PIXELS VALUE IS LESS THAN THE THRESHOLD T
                  pic_arr2[i,j,channel]=0   #SET TO 0
              else:
                  pic_arr2[i,j,channel]=255 #SET TO 255
  plt.imshow(pic_arr2)
````

````python
#greyscale
pic_arr2=gray_scale.copy()
for i in range(0,height):   #FOR EVERY ROW   
        for j in range(0,width):   #FOR EVERY COLUMN
            if pic_arr2[i,j]<T: #IF THE PIXELS VALUE IS LESS THAN THE THRESHOLD T
                pic_arr2[i,j]=0   #SET TO 0
            else:
                pic_arr2[i,j]=255 #SET TO 255
plt.imshow(pic_arr2, cmap='gray')
````
